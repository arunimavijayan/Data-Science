```python
#1. Write a Python Program to Add Two Matrices?
n=int(input("Enter n in an nxn matrix: "))
a=[]
b=[]
add_res=[]
#Taking input----------------------
for i in range(n):
    row=[]    
    for j in range(n):
        row.append(int(input()))
    a.append(row)
print(a)

for i in range(n):
    row=[]    
    for j in range(n):
        row.append(int(input()))
    b.append(row)
print(b)

for i in range(n):
    row=[]    
    for j in range(n):
        row.append(0)
    add_res.append(row)

#Displaying Matrix----------------------
print("FIRST MATRIX")
for i in range(n):
    for j in range(n):
        print(a[i][j], end=" ")
    print()

print("SECOND MATRIX")
for i in range(n):
    for j in range(n):
        print(b[i][j], end=" ")
    print()
    
#Addition--------------------------
for i in range(n):
    for j in range(n):
        add_res[i][j]=a[i][j]+b[i][j]

print("Resultant Matrix :\n")
for i in range(n):
    for j in range(n):
        print(add_res[i][j], end=" ")
    print()
```

```python
#2. Write a Python Program to Multiply Two Matrices?
n=int(input("Enter n in an nxn matrix: "))
a=[]
b=[]
mul_res=[]
#Taking input----------------------
for i in range(n):
    row=[]    
    for j in range(n):
        row.append(int(input()))
    a.append(row)
print(a)

for i in range(n):
    row=[]    
    for j in range(n):
        row.append(int(input()))
    b.append(row)
print(b)

for i in range(n):
    row=[]    
    for j in range(n):
        row.append(0)
    mul_res.append(row)

#Displaying Matrix----------------------
print("FIRST MATRIX")
for i in range(n):
    for j in range(n):
        print(a[i][j], end=" ")
    print()

print("SECOND MATRIX")
for i in range(n):
    for j in range(n):
        print(b[i][j], end=" ")
    print()
    
#Multiplication--------------------------
for i in range(len(a)):
    for j in range(len(b[0])):
        for k in range(len(b)):
            mul_res[i][j]+=a[i][k]*b[k][j]

              
print("Resultant Matrix :\n")
for i in range(n):
    for j in range(n):
        print(mul_res[i][j], end=" ")
    print()
```

```python
#3. Write a Python Program to Transpose a Matrix?
n=int(input("Enter n in an nxn matrix: "))
a=[]
b=[]
#Taking input----------------------
for i in range(n):
    row=[]    
    for j in range(n):
        row.append(int(input()))
    a.append(row)
for i in range(n):
    row=[]    
    for j in range(n):
        row.append(0)
    b.append(row)
#Displaying Matrix----------------------
print("MATRIX")
for i in range(n):
    for j in range(n):
        print(a[i][j], end=" ")
    print()
#Transpose-----------------------------
for i in range(n):
    for j in range(len(a[0])):
        b[i][j]=a[j][i]
print("Transposed Matrix: ")        
for i in range(n):
    for j in range(n):
        print(b[i][j], end=" ")
    print()
```

```python
#5. Write a Python Program to Remove Punctuation From a String?

punc='''!()-[]{};:'"\,<>./?@#$%^&*_~'''
no_punc=""
sent=input()
for i in sent:
    if i not in punc:
        no_punc+=i

print(no_punc)
```
